```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS
```

```python
def getContents(text):
    title = wikipedia.search(text)[0]
    page = wikipedia.page(title)
    return(page.content)
    
    
```

```python
query = "Data Science"
wikipedia.search(query)
```

```python
wikipedia.search(query)[0]
```

```python
text = getContents(query)
text
```

```python
def createWordCloud(text):
    stopwords = STOPWORDS
    
    wc = WordCloud(background_color ='white', max_words = 100,stopwords =stopwords,width = 1200, height = 900)
    
    wc.generate(text)
    
    #create the image file from the wordcloud
    filename = "Datascience.png"
    wc.to_file(filename)
    return(filename)
    
    
```

```python
f = createWordCloud(text)
```

```python
print(f)
```
